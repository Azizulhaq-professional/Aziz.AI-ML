### Project Title: Visualizing Sales Trends with Matplotlib Line Plot
#### Objective:
##### Understand and practice all major parameters of the plt.plot() function in Matplotlib by visualizing sample sales data.

```python
# Step 1: Setup
import matplotlib.pyplot as plt
import numpy as np
```

##### Explanation:
- matplotlib.pyplot is imported as plt (conventionally).

- numpy is used to generate sample data.

```python
# Step 2: Generate Sample Data
months = np.arange(1, 13)
sales = np.array([2500, 2700, 3000, 3200, 3500, 3700, 4000, 3800, 3600, 3900, 4200, 4500])
print(months)
print(sales)
```

#### Explanation:
- months: 1 to 12 (representing Jan to Dec)

- sales: Monthly sales in arbitrary currency units

```python
#  Step 3: Basic Line Plot
plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

#### Explanation:
- A simple line chart

- grid(), xlabel(), ylabel() for labeling and improving readability

```python
#  Step 4: Use of color, linestyle, linewidth
plt.plot(months, sales, color='green', linestyle='--', linewidth=2)
plt.title("Monthly Sales - Line Style")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

#### Explanation:
- color='green': line color

- linestyle='--': dashed line

- linewidth=2: thicker line

```python
# Step 5: Use of marker, markersize, markerfacecolor, markeredgecolor
plt.plot(months, sales, marker='o', markersize=8, markerfacecolor='red', markeredgecolor='black')
plt.title("Monthly Sales - Markers")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

#### Explanation:
- marker='o': circular marker

- markersize=8: size of the marker

- markerfacecolor, markeredgecolor: color customization

```python
# Step 6: Use of label for Legend
plt.plot(months, sales, label='Sales 2025', color='blue')
plt.title("Monthly Sales with Legend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

#### Explanation:
- label='Sales 2025': legend entry

- plt.legend(): shows the legend on the graph

```python
# Step 7: Use of alpha, zorder
plt.plot(months, sales, color='orange', alpha=0.6, zorder=3)
plt.title("Monthly Sales - Transparency and Layering")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True, zorder=0)
plt.show()
```

#### Explanation:
- alpha=0.6: transparency (1 = opaque, 0 = transparent)

- zorder=3: controls layer order (higher = on top)

```python
# Step 8: Use of drawstyle
plt.plot(months, sales, drawstyle='steps-post', color='purple')
plt.title("Monthly Sales - Step Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

#### Explanation:
- drawstyle='steps-post': makes the line step-shaped

- Options: 'default', 'steps-pre', 'steps-mid', 'steps-post'

```python
# Step 9: Combining Multiple Lines with Different Styles
sales_2024 = sales - 300

plt.plot(months, sales, label='2025', color='green', marker='o')
plt.plot(months, sales_2024, label='2024', color='blue', linestyle='--', marker='x')
plt.title("Sales Comparison 2024 vs 2025")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step 10: Add Custom Ticks
plt.plot(months, sales, color='brown', marker='s')
plt.title("Custom Ticks on X-axis")
plt.xticks(ticks=months, labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.yticks(np.arange(2000, 5001, 500))
plt.grid(True)
plt.show()
```
