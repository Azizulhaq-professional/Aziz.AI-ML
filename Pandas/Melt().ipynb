### Pandas melt function is used to transform or reshape data.

#### Syntax: pandas.melt()
#### pd.melt()
### Parameters:
- frame
- var_name
- id_vars
- value_name
- value_vars

## Pandas Melt

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\DELL XPS\Desktop\AI & ML\Data\Movie_ratings.csv')
df
```

```python
pd.melt(df)
```

```python
pd.melt(df, id_vars = ['Genre'])
```

```python
pd.melt(df, id_vars = ['Year'])
```

### Unpivot
#### A dataframe turn it from a wide format(many columns) to a long format (few columns but many rows)

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Genre'])
```

```python
df
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Rating%'])
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Rating%'], var_name = 'Category')
```

```python
pd.melt(df, id_vars = ['Year'], value_vars = ['Rating%'], var_name = "Category", value_name = 'Data')
```
