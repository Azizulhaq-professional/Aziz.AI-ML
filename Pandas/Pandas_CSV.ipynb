### Handling Missing Values in Python dropna method

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\DELL XPS\\Desktop\\AI & ML\\Data\\Fortune_isnull.csv')
df
```

```python
df.dropna() # it will drop the missing values of the dataset
```

```python
# df.dropna() we check the axis parameters
df.dropna(axis = 1)
```

```python
# df.dropna() we check the hoq parameters
df.dropna(axis=1, how = 'any')
# it is showing errors because we didn't give a axis parameters because it value is none by default
```

```python
df.dropna(how = 'all')
```

```python
#Now we check the threshold parameters
df.dropna(thresh = 4)
```

```python
#Now we check the subset parameters
df.dropna(subset = ['ID'])
```

```python
df.dropna(subset = ['Profit'])
```

```python
#Now we use inplace parameters
df = pd.read_csv('C:\\Users\\DELL XPS\\Desktop\\AI & ML\\Data\\Fortune_isnull.csv')
df
```

```python
df.dropna(inplace = True) # drop rows with mission values
df
```
