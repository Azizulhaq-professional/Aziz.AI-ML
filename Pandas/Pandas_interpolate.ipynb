## What is interpolate() function is basically used to fill NaN values in data frame or Series.
- Very powerful function
- Uses various interpolation techniques
### Syntax: DataFrame.interpolate()
###         Series.interpolate
***Parameters: method, inplace, axis, limit_direction, limit, limit_area***
### Pandas interpolate() function
