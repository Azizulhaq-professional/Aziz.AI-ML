# Pandas concate() Function in Python


```python
import pandas as pd
```

```python
sr1 = pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5,6,7])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4],    #DataFrame Data
                  'Name': ['A', 'B', 'C','D'],
                  'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID': [5,6,7,8],
                  'Name': ['E', 'F', 'G', 'H'],
                  'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2, df1], axis = 1)
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4],
                  'Name': ['A', 'B', 'C','D'],
                  'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID': [3,4],
                  'Name': ['C','D'],
                  'Class': [7,8]})
df2
```

```python
df3 = pd.DataFrame({'ID': [5,6], 'Name': ['E','F'], 'Class': [8,9]})
df3
```

```python
pd.concat([df1, df2, df3], axis = 1)
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
import pandas as pd

# Concatenating dataframes df1, df2, and df3 along columns with df1's index alignment
pd.concat([df1, df2, df3], axis=1, join='outer')  # align on the index
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4],
                  'Name': ['A', 'B', 'C','D'],
                  'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID': [5,6,7,8],
                  'Name': ['E', 'F', 'G', 'H'],
                  'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
pd.concat([df1, df2], keys = ['df1','df2'])
```

```python
pd.concat([df1, df2], keys = ['First df','Second df']) #without axis parameter it will give rows wise keys names
```

```python
pd.concat([df1, df2], axis = 1,  keys = ['First df','Second df']) #with axis parameters it will give column key names
```

```python
df2 = pd.DataFrame({'Marks': [40, 63, 91, 34]})
df2
```
