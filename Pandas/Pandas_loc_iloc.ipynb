```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Class': ['A', 'B', 'A', 'B'],
    'Age': [20, 21, 22, 23]
}
df = pd.DataFrame(data)

# Check column names
print(df.columns)

# Clean column names (remove spaces)
df.columns = df.columns.str.strip()

# Access the data
print(df.loc[0:3, 'Class'])
```

# Pandas iloc
***integer location-based indexing***

**Syntax:** DataFrame.iloc

```python
df.iloc[0]
```

```python
df.iloc[:, 0]
```

```python
df.iloc[[0,1]] # for list of labels
```

```python
# Assuming df has 11 rows, you need a boolean list of length 11
bool_indexer = [True, False, True, False, False, True, False, True, False, False, True]
df.loc[bool_indexer]
```

```python
# For selecting rows at specific positions
df.iloc[[0, 2]]  # This will select the rows at positions 0 and 2
```
