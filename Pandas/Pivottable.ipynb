### The Levels in the pivot table will be stored in MultiIndex objects (hirarchical indexes) on the index and columns of the result DataFrame.

#### Syntax: DataFrame.pivot_table()

#### Parameters:

- values
- index
- fill_value
- margins-
- columns
- dropna
- aggfunc

## Pivot Table

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\DELL XPS\Desktop\AI & ML\Data\Movie_ratings.csv')
df
```

```python
df.pivot_table(index = 'Genre')  # By default we get mean value
```

```python
df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum')
```

```python
df
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = 'none')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = '0')
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', fill_value = '0', margins = True)
```

```python
# DropNa parameters actumatically drop Nan Values
df.pivot_table(index = 'Genre', columns = 'Year', dropna = True)
```

```python
df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum', fill_value = '0', margins = True)
```
